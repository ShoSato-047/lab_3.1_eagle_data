#### Problem 1

Before creating clean/DRY solutions, let's make the unclean solution.  For this problem, create the WET solution using a hefty amount of copy-and-paste.

```python
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
eagles = pl.read_csv('./data/bald_eagle_subsample.csv')
eagles.head()
```

```python
# Your copy-and-paste/WET solution here.
eagles_NEW = (eagles
             .with_columns(
                           pl.col('KPH').sqrt().alias('KPH'),
                           pl.col('Sn').sqrt().alias('Sn'),
                           pl.col('AGL0').sqrt().alias('AGLO'),
                           pl.col('abs_angle').sqrt().alias('abs_angle'),
                           pl.col('absVR').sqrt().alias('absVR')
                          )
              .with_columns(
                            transformed_KPH = (pl.col('KPH') - pl.col('KPH').mean()) / pl.col('KPH').std(),
                            transformed_Sn = (pl.col('Sn') - pl.col('Sn').mean()) / pl.col('Sn').std(),
                            transformed_AGL0 = (pl.col('AGL0') - pl.col('AGL0').mean()) / pl.col('AGL0').std(),
                            transformed_VerticalRate = (pl.col('VerticalRate') - pl.col('VerticalRate').mean()) / pl.col('VerticalRate').std(),
                            transformed_abs_angle = (pl.col('abs_angle') - pl.col('abs_angle').mean()) / pl.col('abs_angle').std(),
                            transformed_absVR = (pl.col('absVR') - pl.col('absVR').mean()) / pl.col('absVR').std()
                           )
            )
eagles_NEW.head()
```

```python
# eagles_NEW.columns
```

```python
transformed_cols = ['transformed_KPH',
                    'transformed_Sn',
                    'transformed_AGL0',
                    'transformed_VerticalRate',
                    'transformed_abs_angle',
                    'transformed_absVR']
eagles_stacked = (eagles_NEW
                 .unpivot(transformed_cols, 
                          index='Animal_ID', 
                          variable_name='variables', 
                          value_name='values')
                )
eagles_stacked
```

```python
g = sns.FacetGrid(eagles_stacked, col='variables', col_wrap=3)
g.map(sns.histplot, 'values')

# Set the y-axis limit
plt.ylim(0, 700) 
```

```python
columns_to_transform = [ 'KPH','Sn','AGL0','abs_angle','absVR', 'VerticalRate']
```

#### Problem 3

First, complete the task using a `dict` comprehension to make a `dict` of column name/expression pairs.  Apply the expressions by unpacking the `dict` using `**` inside `with_columns`.

```python
# Create a dict comprehension to apply the sqrt transformation

sqrt_transforms = {col: pl.col(col).sqrt().alias(col) for col in columns_to_sqrt}
```

```python
# Create another dict comprehension for the standardization process

standardize_transforms = {f"transformed_{col}": (pl.col(col) - pl.col(col).mean()) / pl.col(col).std() 
                           for col in columns_to_transform}
```

```python
# Unpack the dictionaries inside the with_columns

eagles_NEW = (eagles
              .with_columns(**sqrt_transforms)  # Unpack the sqrt transformations
              .with_columns(**standardize_transforms)  # Unpack the standardization
             )

eagles_NEW.head()
```

#### Problem 4

Now refactor the expressions in the last two problems by packing (A) the column expressions, and (B) the comprehensions in a `lambda` function with a relevant name.

**Note.** Normally this refactoring would take place in the original code cells, but here I want you to
1. Copy your code for each part into a new cell below,
2. Define your `lambda` helper functions above the main data management dot-chain (same cell), and
3. Perform your refactoring of the data management dot-chain (also in this new cell).

```python
# (A) Lambda function to apply square root transformation to selected columns
sqrt_transformer = lambda cols: {col: pl.col(col).sqrt().alias(col) for col in cols}

# (B) Lambda function for standardizing columns (mean and standard deviation)
standardize_transformer = lambda cols: {f"transformed_{col}": (pl.col(col) - pl.col(col).mean()) / pl.col(col).std() for col in cols}

# List of columns to apply sqrt and standardize
sqrt_columns = ['KPH', 'Sn', 'AGL0', 'abs_angle', 'absVR']
standardize_columns = sqrt_columns + ['VerticalRate']

# Apply the transformations using the lambda functions
eagles_NEW = (eagles
              .with_columns(**sqrt_transformer(sqrt_columns))  # Apply sqrt transformation
              .with_columns(**standardize_transformer(standardize_columns))  # Apply standardization
             )

eagles_NEW.head()
```

```python
# Code for Stack/Transform/Unstack here

transformed_cols = [f"transformed_{col}" for col in standardize_columns]

eagles_stacked = (eagles_NEW
                 .unpivot(transformed_cols, 
                          index='Animal_ID', 
                          variable_name='variables', 
                          value_name='values')
                )
eagles_stacked
```
